```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
# Load Dataset
data = pd.read_excel("EastWestAirlines.xlsx", sheet_name="data")
```

```python
data.shape# rows and columns
```

```python
# Preprocessing
# Drop non-numeric or ID columns
if 'ID#' in data.columns:
    data = data.drop('ID#', axis=1)
```

```python
#check missing values
data.isnull().sum() #There is 0 missing values
```

```python
#check duplicates
data.duplicated().sum() #There is 1 duplicates
```

```python
#dropping duplicate
data = data.drop_duplicates()
```

```python
data.describe()
```

```python
data.boxplot() ##We can visualize outliers using boxplot
```

```python
def outlier_capping(df2, col): #outlier_apping using Quantiles and Inter Quantile range
    Q1, Q3 = df2[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5  * IQR
    df2[col] = df2[col].clip(lower, upper)   # directly cap values using clip()

for col in data.select_dtypes(include=['int', 'float']).columns:
    outlier_capping(data, col)
```

```python
data.boxplot()#no outliers
```

```python
std_sca=StandardScaler() ## Calling standardscalar() package using std_sca
data1=pd.DataFrame(std_sca.fit_transform(data),columns=data.columns) #fitting our dataset columns for transforming into standard format
data1.head()
```

```python
### Implement of PCA
from sklearn.decomposition import PCA ##importing PCA from sklearn package
```

```python
pca=PCA(n_components=7) ##top 10 columns is n_components=10
pca.fit(data1) ## fit our standardized data in pca()
```

```python
New_data=pca.transform(data1) #now, transforming our fitted PCA data
New_data.shape
```

```python
'''### PCA Conclusion
Principal Component Analysis (PCA) was applied to reduce the dimensionality of the scaled data while retaining most of the variance.
 By reducing the number of features, PCA helps to simplify the clustering process and potentially improve the performance of clustering algorithms by removing noise and multicollinearity.
 In this analysis, the data was reduced to 7 principal components.'''
```

```python
# ---K-Means Clustering ---
inertia = []
K = range(2,11)
for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(New_data)
    inertia.append(km.inertia_)
```

```python
plt.plot(K, inertia, 'bx-')
plt.xlabel("k")
plt.ylabel("Inertia (SSE)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
# Fit KMeans with chosen k (say k=5 from elbow)
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(New_data)
print("K-Means Silhouette Score:", silhouette_score(New_data, kmeans_labels))
```

```python
''' ### K-Means Clustering Conclusion
K-Means clustering was performed on the PCA-transformed data. The elbow method was used to determine an optimal number of clusters (k).
 The silhouette score for the chosen k was calculated to evaluate the clustering performance. The visualization shows the data points colored according to their assigned clusters, providing a visual representation of the K-Means output.
 The interpretation of these clusters would involve examining the characteristics of the data points within each cluster (e.g., average balance, miles, etc.).'''
```

```python
# ---  Hierarchical Clustering ---
linked = linkage(data1, method='ward')
plt.figure(figsize=(10,6))
dendrogram(linked, truncate_mode='level', p=5)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters=5, linkage='ward') ## Hierarchical clustering silhouette score
hc_labels = hc.fit_predict(New_data)
print("Hierarchical Silhouette Score:", silhouette_score(New_data, hc_labels))
```

```python
# ---  DBSCAN ---
db = DBSCAN(eps=2, min_samples=5)
db_labels = db.fit_predict(New_data)

# Check number of clusters found (ignoring noise = -1)
n_clusters_db = len(set(db_labels)) - (1 if -1 in db_labels else 0)
print("DBSCAN clusters:", n_clusters_db)
```

```python
if n_clusters_db > 1:
    print("DBSCAN Silhouette Score:", silhouette_score(data1, db_labels))
```

```python
''' ### DBSCAN Clustering Conclusion
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) was applied to the PCA-transformed data.
Unlike K-Means, DBSCAN does not require specifying the number of clusters beforehand and can identify clusters of arbitrary shapes, as well as noise points.
The number of clusters found and the silhouette score (if applicable) were reported.
The visualization shows the clusters identified by DBSCAN, including any points labeled as noise (-1).
The interpretation of DBSCAN results focuses on the density-based grouping of data points.'''
```

### Cluster Analysis & Interpretation
- **KMeans**: Produces well-separated spherical clusters. Silhouette Score helps confirm cluster quality.
- **Hierarchical**: Dendrogram helps visualize merging process. Useful for understanding data structure.
- **DBSCAN**: Detects arbitrary shapes and noise. Good for datasets with non-spherical clusters.
# Summary:
Each algorithm has strengths. KMeans works best when clusters are compact, Hierarchical is good for nested structures, and DBSCAN is robust to noise and irregular shapes.
